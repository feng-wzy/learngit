```python
import pandas as pd
import datetime
from pandas import Series,DataFrame

def days(max_time_str,min_time_str):
    date1=datetime.datetime.strptime(max_time_str[0:10],"%Y/%m/%d")
    date2=datetime.datetime.strptime(min_time_str[0:10],"%Y/%m/%d")
    num=(date1-date2).days
    return num

def months(max_time_str,min_time_str):
    year1=datetime.datetime.strptime(max_time_str[0:10],"%Y/%m/%d").year
    year2=datetime.datetime.strptime(min_time_str[0:10],"%Y/%m/%d").year
    month1=datetime.datetime.strptime(max_time_str[0:10],"%Y/%m/%d").month
    month2=datetime.datetime.strptime(min_time_str[0:10],"%Y/%m/%d").month
    num=(year1-year2)*12+(month1-month2)
    return num

file_pos = "C://Users//24804//Desktop//data1.xlsx"
sn = pd.ExcelFile(file_pos)
# 删除D等级
# del_list = ['E36', 'E52', 'E82', 'E99', 'E100', 'E101', 'E102', 'E103', 'E107', 'E108', 'E109', 'E111', 'E112', 'E113', 'E114', 'E115', 'E116', 'E117', 'E118', 'E119', 'E120', 'E121', 'E122', 'E123']
# 分别保存的进项和销项统计数据
writer1 = pd.ExcelWriter('C://Users//24804//Desktop//post//result.xlsx')
# 遍历所有sheet数据


new_data1 = {'企业代码':[],'进项发票作废率':[],'进项负数发票率':[],'进项每月平均交易额':[],'进项每月交易次数':[]}
new_data2 = {'企业代码':[],'销项发票作废率':[],'销项负数发票率':[],'销项每月平均交易额':[],'销项每月交易次数':[],'销售收入增长率':[]}
df1 = DataFrame(new_data1)
df2 = DataFrame(new_data2)
flag = True
insert_datas = []
for xsn in sn.sheet_names[1:]:
    # 读取文件
    datas = pd.read_excel(file_pos, sheet_name=xsn)
    # 删除D等级
    # datas = datas[~datas['code'].isin(del_list)]
    # datas['date']=pd.to_datetime(datas['date'],format='%Y/%m/%d')
    # tmp_datas['date'] = [x.strftime('%Y/%m/%d') for x in tmp_datas['date']]
    # datas['diff_date'] = datas['date'].diff();
    # datas.replace({'NaT': '0 day'}, inplace=True)
    # datas.set_index('date', drop=True)
    # 找到全部公司名称代号
    code_list = list(set(list((datas['code']))))
    for name in code_list:
        tmp_datas = datas[datas['code'] == name]
        tmp_datas.index = range(len(tmp_datas))
        # tmp_datas = tmp_datas.groupby(tmp_datas['date']).sum()
        insert_datas.append(name)
        # 作废数
        cacel_count = len(tmp_datas[tmp_datas['tax_status'] == 'b'])
        # 有效数
        valid_count = len(tmp_datas[tmp_datas['tax_status'] == 'a'])
        # 发票作废率
        count1 = (cacel_count / (cacel_count + valid_count))*100
        # 负数发票数
        neg_count = len(tmp_datas[tmp_datas['cost'] < 0])
        # 负数发票率
        count2 = (neg_count / valid_count) * 100
        
        # 销售收入（销项）
        # if xsn == 'sold_info':
        	# income = 
        
        # 转换时间
        tmp_datas['date'] = [x.strftime('%Y/%m/%d') for x in tmp_datas['date']]
        # 时间最大值
        max_time = tmp_datas.iloc[0:,1].max()
        # 时间最小值
        min_time = tmp_datas.iloc[0:,1].min()
        # 时间差
        diff_time = months(max_time, min_time) + 1
        # 有效票
        valid_tax = tmp_datas[tmp_datas['tax_status'] == 'a']
        # 平均月交易额
        avg_money = valid_tax['totle_cost'].sum() / diff_time
        # 平均每月交易次数
        # trans_sum_count = len(tmp_datas)
        trans_count = len(tmp_datas) / diff_time
        insert_datas += [count1, count2, avg_money, trans_count,]
        if flag:
            df1.loc[len(df1)] = insert_datas
            df1.to_excel(writer1,sheet_name='进项信息',index=False)
        else:
            # 合并日期
            merge_time = tmp_datas.groupby(tmp_datas['date']).sum()
            # 销售收入增长率
            income_info= list((merge_time['cost'] - merge_time['cost'].shift(1)).fillna(1))
            diff_time_day = days(max_time,min_time)
            income_tax = (sum(income_info) / diff_time_day)*100
            insert_datas.append(income_tax)
            df2.loc[len(df2)] = insert_datas
            df2.to_excel(writer1,sheet_name='销项信息',index=False)
        insert_datas = []
    flag = False
writer1.save()
```
